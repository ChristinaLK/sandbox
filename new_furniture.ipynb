```python
from ipythonblocks import *

room_width=int(84+54)
room_depth=int(45+90)

#things on the walls
window_distance = 46
window = 45
mirror_distance = 64
mirror = 28

#doors
door = 13
closets = 10
door_w = 41

#furniture
bed_w = 43
bed_l = 84
small_table_w = 17
small_table_d = 15
big_table_w = 25
big_table_d = 19
desk_w = 48
desk_d = 24
shelf_w = 25
shelf_d = 12
drawers_w = 17
drawers_d = 24
```

```python
# create room
grid = BlockGrid(room_width, room_depth, block_size=5, fill=colors['LightBlue'], lines_on = False)

## furniture
# bed
grid[0+1:bed_w+1,-bed_l:-1] = colors['Blue']
#small table
grid[bed_w+2:bed_w+small_table_w+2,-small_table_d:-1] = colors['Brown']
# drawers
grid[bed_w+small_table_w+3:bed_w+small_table_w+drawers_w+3,-drawers_d:-1] = colors['White']
grid[bed_w+small_table_w+drawers_w+3:bed_w+small_table_w+drawers_w+drawers_w+3,-drawers_d:-1] = colors['White']
#shelf
grid[0:shelf_d,18:18+shelf_w] = colors['White']
#desk
grid[30:30+desk_w,0:desk_d] = colors['Black']

# add doors / windows/ stuff on the walls
grid[-(door+30):-door,0] = colors['White']
grid[0,window_distance:window_distance+window] = colors['White']
grid[mirror_distance:mirror_distance+mirror,-1] = colors['Black']
grid[-closets:,-6:-1] = colors['White']
grid[-closets:,80:85] = colors['White']
grid[-closets:,86:91] = colors['White']
grid[-closets:,41:46] = colors['White']
marker = 0
slant = door
for d in list(range(door))[1:]:
    grid[-slant:,marker:marker+3] = colors['White']
    marker = marker +3
    slant = slant - 1

grid
```
