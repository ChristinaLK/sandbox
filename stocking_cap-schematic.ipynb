```python
from ipythonblocks import BlockGrid
from ipythonblocks import colors
```

```python
knit = colors['IndianRed']
purl = colors['DarkRed']

width = 46
center = int(width/2)
rows_1 = 12
rows_2 = 30
rows_3 = 84
rows_total = rows_1+rows_2+rows_3

hat = BlockGrid(width, rows_total, fill=colors['LightGray'], block_size=10)

hat[0:rows_1,:] = knit
hat[rows_1:rows_1+rows_2,:] = knit

hw = center -1 
for row in range(rows_3):
    hat[rows_1+rows_2+row,center-hw:center+hw] = knit
    if (row + 1) % 4 == 0:
        hw = hw - 1
```

```python
# rib
for rib in range(width):
    if rib % 4 == 0:
        hat[0:rows_1,rib] = purl
        hat[0:rows_1,rib+1] = purl

# non-decreasing spiral
spiral_length = 17
start_point = 0

for row in range(rows_2):
    hat[rows_1+row,start_point:start_point+spiral_length] = purl
    start_point = start_point + 1
    if row % 2 == 1:
        spiral_length = spiral_length - 1

double_rows = 44
single_rows = rows_3 - double_rows
neutral_space = 0
for row in range(rows_3):
    if row % 4 == 0:
        neutral_space = neutral_space + 1
    if (start_point+spiral_length-1) >= (width - neutral_space):
        start_point = neutral_space
    if(row == double_rows):
        spiral_length = spiral_length - 1
    hat[rows_1+rows_2+row,start_point:start_point+spiral_length] = purl
    start_point = start_point + 1

hat
```
