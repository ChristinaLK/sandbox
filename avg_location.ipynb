```python
location_dict = {"Chicago": (41.88, -87.62),
                 "Ambler": (39.95, -75.17),
                 "San Antonio": (29.41, -98.5),
                 "El Segundo": (34.05, -118.25),
                 "St Olaf": (44.46, -93.16),
                 "Salem":(44.93, -123.03),
                 "Budapest":(47.47, 19.05),
                 "DC":(38.09, -77.04),
                 "Vancouver":(49.25,-123.1)
                 }
```

```python
time_dict = {"Chicago": 0,
                 "Ambler": 0,
                 "San Antonio": 0,
                 "El Segundo": 0,
                 "St Olaf": 0,
                 "Salem": 0,
                 "Budapest": 0,
                 "DC": 0,
                 "Vancouver": 0,
             }
```

```python
def total_months():
    total = 0.0
    for location in time_dict:
        total += time_dict[location]
    return total
```

```python
def avg_location():
    months = total_months()
    tuple_sum = [0.0,0.0]
    if months == 0:
        return tuple_sum
    else: 
        for location in location_dict:
            location_weight = time_dict[location]/months
            tuple_sum[0] += location_weight*location_dict[location][0]
            tuple_sum[1] += location_weight*location_dict[location][1]
            #print tuple_sum
        tuple_avg = (tuple_sum[0], tuple_sum[1])
        return tuple_avg
        #print tuple_avg
```

```python
avg_loc = avg_location()
print avg_loc
```

```python
%pylab inline
import matplotlib.pyplot as plt
```

```python
def location_plotter(avg_location):
    for location in location_dict:
        if time_dict[location] != 0:
            plt.plot([avg_location[1],location_dict[location][1]], [avg_location[0],location_dict[location][0]])
    plt.axis([-125, -70, 28, 50])
    plt.show()
```

```python
location_plotter(avg_loc)
```
